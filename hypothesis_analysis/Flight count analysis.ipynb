### Посчитать количество полетов, которые заканчиваются в одном, а начинаются в другом 

```python
def count_not_normal_flights(dataframe):
    prev = ''
    count = 0
    for elem in dataframe.values:
        if pd.isna(elem[3]):
            continue
        if pd.isna(elem[4]):
            prev = ''
            continue
        if prev != '' and elem[3].lower() != prev.lower():
            count += 1
        prev = elem[4]
    return count]
```

### Убираем клиентов, которые летали меньше 3х раз

```python
df = df[df['flights'] > 3]
```

```python
df
```

```python
value = 10000
unique_names = df.name.unique()
for i in range (len(unique_names)):
    name = unique_names[i]
    sub_df = df[df['name'] == name].sort_values(by='depart_date')
    not_ok = count_not_normal_flights(sub_df)
    flights = sub_df.values[0][-1]
    not_normal_flights.loc[i] = [name] + [not_ok] + [flights]
    if i == value:
        print(i)
        value += 10000
```

### Создадим словарь: Код аэропорта - страна

```python
reg_flights = pd.read_csv('data/Registered Flights.csv', index_col='id')
```

```python
reg_flights
```

```python
airport_dict = {}
value = 10000
for i in range(len(reg_flights)):
    d_country = reg_flights.iloc[i]['Depart_Country']
    d_air = reg_flights.iloc[i]['Depart_Airport']
    a_country = reg_flights.iloc[i]['Arrival_Country']
    a_air = reg_flights.iloc[i]['Arrival_Airport']
    if not pd.isna(d_country):
        if airport_dict.get(d_air) is None:
            airport_dict[d_air] = d_country
    if not pd.isna(a_country):
        if airport_dict.get(a_air) is None:
            airport_dict[a_air] = a_country
    if i == value:
        print(i)
        value += 10000
```

```python
for k, v in airport_dict.items():
    if v.find('United') != -1:
        airport_dict[k] = 'USA'
```

### Считаем международные перелеты

```python
def count_not_cross_country_flights(dataframe):
    prev = ''
    count = 0
    for elem in dataframe.values:
        if elem[8] == '' or elem[9] == '':
            continue
        if elem[8] != elem[9]:
            count += 1
    return count
```

```python
value = 10000
unique_names = df_new.name.unique()
for i in range (len(unique_names)):
    name = unique_names[i]
    sub_df = df_new[df_new['name'] == name].sort_values(by='depart_date')
    cross = count_not_cross_country_flights(sub_df)
    cross_country.loc[i] = [name] + [cross]
    if i == value:
        print(i)
        value += 10000
```
