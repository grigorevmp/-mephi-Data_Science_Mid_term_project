## Выделяем людей, летавших более одного раза

```python
names = []

for i in range(len(boarding_data)):
    names.append(boarding_data['PassengerFirstName'][i] + ' ' + boarding_data['PassengerLastName'][i])
#pd.unique(boarding_data['PassengerFirstName'], boarding_data['PassengerSecondName'], boarding_data['PassengerLastName'])
```

```python
name = pd.DataFrame(names)
dup = name.duplicated(keep='last')
```

```python
index = 0
all_dup_names = []
for i in dup:
    if i:
        all_dup_names.append(names[index])
    index += 1
```

```python
dup_names = list(dict.fromkeys(all_dup_names)) # все дубликаты
```

```python
df = pd.DataFrame(dup_names, columns=['Name'])
```

```python
df.to_csv('dup_names.csv', index=False)
```
