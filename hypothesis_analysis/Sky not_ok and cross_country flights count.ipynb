### Для перевода имени колонок к удобному виду

```python
import re
def camel_to_snake(name):
  name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
  return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()
new_cols = {i:camel_to_snake(i) for i in df1_no_dup.columns}
df1_no_dup.rename(columns=new_cols, inplace=True)
```

```python
df1_no_dup
```

### Считаем все: not_ok, между народные, ненормальные международные и кроссконтинентальные

```python
def count_not_normal_flights(dataframe):
    prev_air = ''
    not_ok = 0
    cross_country = 0
    cross_not_ok = 0
    long = 0
    for elem in dataframe.values:
        from_air = elem[0]
        to_air = elem[1]
        from_coun = elem[2]
        to_coun = elem[3]
        from_con = elem[4]
        to_con = elem[5]
        if pd.isna(from_air):
            continue
        if pd.isna(to_air):
            prev_air = ''
            continue
        if prev_air != '' and from_air.lower() != prev_air.lower():
            not_ok += 1
            if from_coun != '' and to_coun != '':
                if from_coun.lower() != to_coun.lower():
                    cross_not_ok += 1
        prev_air = to_air
        if from_coun != '' and to_coun != '':
            if from_coun.lower() != to_coun.lower():
                cross_country += 1
        if from_con != '' and to_con != '':
            if from_con.lower() != to_con.lower():
                long += 1
    return not_ok, cross_country, cross_not_ok, long
```

### Определяем страны прилета и вылета по коду аэропорта

```python
df1_no_dup['dep_country'] = [country.get(elem,'') for elem in df1_no_dup['from'].values]
```

```python
df1_no_dup['arr_country'] = [country.get(elem,'') for elem in df1_no_dup['to'].values]
```

```python
df1_no_dup['dep_continent'] = [continents.get(elem,'') for elem in df1_no_dup['dep_country'].values]
```

```python
df1_no_dup['arr_continent'] = [continents.get(elem,'') for elem in df1_no_dup['arr_country'].values]
```

### Перевод uid в целочисленный тип

```python
df1_no_dup['uid'] = [int(elem) for elem in df1_no_dup.uid.values]
```

### Создаем таблицу для записи данных

```python
not_normal_flights = pd.DataFrame(columns=['uid', 'name', 'flights', 'not_ok_flights', 'cross_country', 'cross_country_not_ok', 'long_flights'])
```

```python
value = 5000
unique_names = df1_no_dup.uid.unique()
for i in range (len(unique_names)):
    uid = unique_names[i]
    sub_df = df1_no_dup[df1_no_dup['uid'] == uid].sort_values(by='data').reset_index(drop=True)
    name = sub_df['name'][0]
    sub_df_1 = sub_df[['from', 'to', 'dep_country', 'arr_country', 'dep_continent', 'arr_continent']]
    count_flights = len(sub_df)
    not_ok, cross_country, cross_not_ok, long = count_not_normal_flights(sub_df_1)
    not_normal_flights.loc[i] = [uid] +[name] + [count_flights] + [not_ok] + [cross_country] + [cross_not_ok] + [long]
    if i == value:
        print(i)
        value += 5000
```

```python
not_normal_flights
```

### Вычисляем процент подозрительных рейсов и процент международных рейсов

```python
not_normal_flights['percent_not_ok'] = [elem[3] / elem[2] for elem in not_normal_flights.values]
```

```python
not_normal_flights['percent_cross'] = [elem[4] / elem[2] for elem in not_normal_flights.values]
```
